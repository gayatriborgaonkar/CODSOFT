```python
from sklearn.cluster import KMeans
k_rng = range(1,10)
sse=[]

for k in k_rng:
  km = KMeans(n_clusters=k)
  km.fit(iris[['petal_length' , 'petal_width']])
  sse.append(km.inertia_)
```

```python
sse
```

```python
plt.xlabel('k_rng')
plt.ylabel("sum of squared errors")
plt.plot(k_rng,sse)
```

```python
km=KMeans(n_clusters=3, random_state=0)
y_predicted=km.fit_predict(iris[['petal_length', 'petal_width']])
y_predicted
```

```python
from sklearn.metrics import confusion_matrix
from sklearn.cluster import KMeans
from sklearn.preprocessing import LabelEncoder
import seaborn as sns

iris = sns.load_dataset('iris')

kmeans = KMeans(n_clusters=3)
iris['cluster'] = kmeans.fit_predict(iris[['petal_length', 'petal_width']])

label_encoder = LabelEncoder()
iris['species_encoded'] = label_encoder.fit_transform(iris['species'])

cm = confusion_matrix(iris['species_encoded'], iris['cluster'])
print(cm)
```

```python
cm = confusion_matrix(iris['species_encoded'], iris['cluster'])
class_labels=['Setosa','versicolor','virginica']

plt.imshow(cm, interpolation='nearest', cmap=plt.cm.Blues)
plt.title('Confusion Matrix')
plt.colorbar()
tick_marks = np.arange(len(class_labels))
plt.xticks(tick_marks, class_labels)
plt.yticks(tick_marks, class_labels)

for i in range(len(class_labels)):
  for j in range(len(class_labels)):
    plt.text(j,i,str(cm[i][j]), ha='center' ,va='center', color='white')

plt.xlabel('Predicted label')
plt.ylabel('True label')
plt.show()
```
